```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
# trazer a média
df_faturamento["valor"].mean()
```

```python
# gráfico de barras vertical

# converter strings para o formato de data (datetime)
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])

# criar uma coluna apenas com o nome do mês
df_faturamento['mes'] = df_faturamento['data_ref'].dt.month_name()

df_faturamento.plot.bar(x='mes', y='valor')
plt.ticklabel_format(style='plain', axis='y') # remove a notação científica 

plt.title('Faturamento por Mês')
plt.xlabel("Meses")
plt.ylabel("Valores")
plt.tight_layout()
```

```python
# gráfico de linhas
df_faturamento.plot.line(x='mes', y='valor')
plt.ticklabel_format(style='plain', axis='y') 
plt.title('Faturamento por Mês')
plt.xlabel("Meses")
plt.ylabel("Valores")
plt.tight_layout()
```
