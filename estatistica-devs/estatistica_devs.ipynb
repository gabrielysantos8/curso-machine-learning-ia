```python
import pandas as pd
```

```python
pd.__version__
```

```python
dict_medidas = {'idade': [16, 17, 19, 22, 22, 40, 46, 55, 60, 81], 
'altura': [155, 160, 164, 168, 170, 172, 172, 178, 180, 181]}
```

```python
dict_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

## Medidas de posição

```python
# média
df_medidas["idade"].mean()
```

```python
# mediana
df_medidas["idade"].median()
```

```python
#moda
df_medidas["idade"].mode()
```

```python
df_medidas["altura"].mode()
```

## Medidas de dispersão

```python
# variância
df_medidas.idade.var()
```

```python
# desvio padrão
df_medidas.idade.std()
```

```python
# coeficiente de variação
df_medidas.idade.std() / df_medidas.idade.mean() * 100
```

```python
# coeficiente de variação (altura)
df_medidas.altura.std() / df_medidas.altura.mean() * 100
```

## Medidas de forma

```python
# assimetria (idade)
df_medidas.idade.skew()
```

```python
# assimetria (altura)
df_medidas.altura.skew()
```

```python
# curtose (idade)
df_medidas.idade.kurtosis()
```

```python
# curtose (altura)
df_medidas.altura.kurtosis()
```

```python
# obter medidas estatisticas
df_medidas.idade.describe()
```

## Correlação

```python
# correlação do DataFrame (pearson)
df_medidas.corr(method='pearson')
```

```python
# correlação entre 2 variáveis
df_medidas.idade.corr(df_medidas.altura)
```

```python
# correlação do DataFrame (spearman)
df_medidas.corr(method='spearman')
```

## Gráficos

```python
# histograma (idade)
df_medidas.idade.hist()
```

```python
# histograma (altura)
df_medidas.altura.hist()
```

### DataFrame de vendas

```python
dict_vendas = {'categoria': ['feminino', 'masculino', 'infantil', 'casa'],
               'valor': [400000, 600000, 250000, 580000], 'quantidade': [3000, 5000, 1500, 2500]}
```

```python
dict_vendas
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
```

```python
df_vendas
```

```python
# gráfico de barras vertical (valor)
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
# gráfico de barras vertical (valor - ordenado)
df_vendas.sort_values('valor').plot.bar(x='categoria', y='valor')
# df_vendas.sort_values('valor', ascending=False).plot.bar(x='categoria', y='valor')
```

```python
# gráfico de barras horizontal (quantidade)
df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
# gráfico de barras horizontal (quantidade - ordenado)
df_vendas.sort_values('quantidade').plot.barh(x='categoria', y='quantidade')
```

```python
# gráfico de dispersão
df_medidas.plot.scatter(x='idade', y='altura')
```

```python
# gráfico de bloxploat (idade)
df_medidas.idade.plot.box()
```

```python
# gráfico de bloxploat (altura)
df_medidas.altura.plot.box()
```
