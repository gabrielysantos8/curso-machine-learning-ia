```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados_vendas = {
    'mes': [
      'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',  'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'
    ],
    'vendas': [
       2000, 2200, 2300, 2500, 2600, 2700,
       2800, 2900, 3000, 3100, 3200, 3300
    ]
}
```

```python
# tratar dados
df_vendas = pd.DataFrame(dados_vendas)
```

```python
# converter os meses para números
# 1. Crie o dicionário com o mapeamento de nome para número
meses_map = {
    'janeiro': 1, 'fevereiro': 2, 'março': 3, 'abril': 4,
    'maio': 5, 'junho': 6, 'julho': 7, 'agosto': 8,
    'setembro': 9, 'outubro': 10, 'novembro': 11, 'dezembro': 12
}

# 2. Converta o texto para minúsculo e aplique o mapeamento
df_vendas['mes'] = df_vendas['mes'].str.lower().map(meses_map)

# 3. Garanta que o resultado seja do tipo inteiro
df_vendas['mes'] = df_vendas['mes'].astype(int)
```

```python
# treinar modelo
from sklearn.linear_model import LinearRegression
```

```python
X = df_vendas[['mes']]
y = df_vendas['vendas']
```

```python
modelo = LinearRegression().fit(X, y)
```

```python
# prever venda de dezembro
previsao = modelo.predict([[12]])
previsao
```
